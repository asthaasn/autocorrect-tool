```python
from textblob import TextBlob

sentence = input("Enter a sentence: ")

words = sentence.split()
corrected_words = []

for word in words:
    corrected = str(TextBlob(word).correct())
    corrected_words.append(corrected)

print("Corrected Sentence:", " ".join(corrected_words))
```

```python
from textblob import TextBlob

def autocorrect_text(text):
    blob = TextBlob(text)
    return str(blob.correct())

def show_suggestions(word):
    blob = TextBlob(word)
    return blob.words[0].spellcheck()

while True:
    print("\n1. Correct Sentence")
    print("2. Get Suggestions")
    print("3. Exit")

    choice = input("Enter choice: ")

    if choice == '1':
        text = input("Enter sentence: ")
        print("Corrected:", autocorrect_text(text))

    elif choice == '2':
        word = input("Enter word: ")
        print("Suggestions:", show_suggestions(word))

    elif choice == '3':
        print("Exiting...")
        break

    else:
        print("Invalid choice")
```
